```python
# Generate bucket dictionaries
results_dict = {}
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced

def model_results(model_name,y_test,y_pred):
    results_dict[model_name] = {
    'balanced_accuracy_score': balanced_accuracy_score(y_test, y_pred),
    'confusion_matrix' : confusion_matrix(y_test, y_pred),
    'classification_report_imbalanced' : pd.DataFrame(classification_report_imbalanced(y_test, y_pred, output_dict=True)).transpose().loc[['high_risk','low_risk']]
    }
```

# Read the CSV and Perform Basic Data Cleaning

```python
# https://help.lendingclub.com/hc/en-us/articles/215488038-What-do-the-different-Note-statuses-mean-

columns = [
    "loan_amnt", "int_rate", "installment", "home_ownership",
    "annual_inc", "verification_status", "issue_d", "loan_status",
    "pymnt_plan", "dti", "delinq_2yrs", "inq_last_6mths",
    "open_acc", "pub_rec", "revol_bal", "total_acc",
    "initial_list_status", "out_prncp", "out_prncp_inv", "total_pymnt",
    "total_pymnt_inv", "total_rec_prncp", "total_rec_int", "total_rec_late_fee",
    "recoveries", "collection_recovery_fee", "last_pymnt_amnt", "next_pymnt_d",
    "collections_12_mths_ex_med", "policy_code", "application_type", "acc_now_delinq",
    "tot_coll_amt", "tot_cur_bal", "open_acc_6m", "open_act_il",
    "open_il_12m", "open_il_24m", "mths_since_rcnt_il", "total_bal_il",
    "il_util", "open_rv_12m", "open_rv_24m", "max_bal_bc",
    "all_util", "total_rev_hi_lim", "inq_fi", "total_cu_tl",
    "inq_last_12m", "acc_open_past_24mths", "avg_cur_bal", "bc_open_to_buy",
    "bc_util", "chargeoff_within_12_mths", "delinq_amnt", "mo_sin_old_il_acct",
    "mo_sin_old_rev_tl_op", "mo_sin_rcnt_rev_tl_op", "mo_sin_rcnt_tl", "mort_acc",
    "mths_since_recent_bc", "mths_since_recent_inq", "num_accts_ever_120_pd", "num_actv_bc_tl",
    "num_actv_rev_tl", "num_bc_sats", "num_bc_tl", "num_il_tl",
    "num_op_rev_tl", "num_rev_accts", "num_rev_tl_bal_gt_0",
    "num_sats", "num_tl_120dpd_2m", "num_tl_30dpd", "num_tl_90g_dpd_24m",
    "num_tl_op_past_12m", "pct_tl_nvr_dlq", "percent_bc_gt_75", "pub_rec_bankruptcies",
    "tax_liens", "tot_hi_cred_lim", "total_bal_ex_mort", "total_bc_limit",
    "total_il_high_credit_limit", "hardship_flag", "debt_settlement_flag"
]

target = ["loan_status"]
```

```python
# Load the data
file_path = 'LoanStats_2019Q1.csv'
df = pd.read_csv(file_path, skiprows=1)[:-2]
df = df.loc[:, columns].copy()

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()

# Remove the `Issued` loan status
issued_mask = df['loan_status'] != 'Issued'
df = df.loc[issued_mask]

# convert interest rate to numerical
df['int_rate'] = df['int_rate'].str.replace('%', '')
df['int_rate'] = df['int_rate'].astype('float') / 100


# Convert the target column values to low_risk and high_risk based on their values
x = {'Current': 'low_risk'}   
df = df.replace(x)

x = dict.fromkeys(['Late (31-120 days)', 'Late (16-30 days)', 'Default', 'In Grace Period'], 'high_risk')    
df = df.replace(x)

df.reset_index(inplace=True, drop=True)

df.head()
```

# Split the Data into Training and Testing

```python
# Dataframe with datatypes and column names
df_dtypes = pd.DataFrame({
    'column_name': df.columns,
    'dtype': df.dtypes
})
df_dtypes.index = pd.Series(range(len(df.dtypes)))
print(df_dtypes.groupby('dtype').count())
non_numerical_dtypes = df_dtypes.loc[(df_dtypes['dtype']=='object')&(df_dtypes['column_name']!='loan_status')]['column_name'].reset_index(drop=True)
non_numerical_dtypes
```

```python
# Create our features
X = df.copy().drop('loan_status',axis=1)
X = pd.get_dummies(X,columns=non_numerical_dtypes)

# Create our target
y = df['loan_status']
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=20)
```

### Balanced Random Forest Classifier

```python
# Resample training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
# StandardScaler instance.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fitting the Standard Scaler with training data
X_scaler = scaler.fit(X_train)
#Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# Create a BalancedRandomForest Classifier
brfc_model = BalancedRandomForestClassifier(n_estimators=150, random_state = 20)
# Fit the model
brfc_model.fit(X_train_scaled, y_train)
# Make predictions using testing data
predictions = brfc_model.predict(X_test_scaled)
```

```python
# Calculated the balanced accuracy score
balanced_accuracy_score(y_test, predictions)
```

```python
# Display the confusion matrix
confusion_matrix(y_test, predictions)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, predictions))
```

```python
# List the features sorted in descending order by feature importance
features = sorted(zip(brfc_model.feature_importances_, X.columns), reverse=True)
features
```

```python
model_results("BalancedRandomForestClassifier",y_test,predictions)
results_dict["BalancedRandomForestClassifier"]['classification_report_imbalanced']
```

### Easy Ensemble AdaBoost Classifier

```python
# Train the EasyEnsembleClassifier
from imblearn.ensemble import EasyEnsembleClassifier
easy_ensemblec_model = EasyEnsembleClassifier(n_estimators=150, random_state = 20)
easy_ensemblec_model.fit(X_train_scaled, y_train)
predictions = easy_ensemblec_model.predict(X_test_scaled)
```

```python
# Calculated the balanced accuracy score
balanced_accuracy_score(y_test, predictions)
```

```python
# Display the confusion matrix
confusion_matrix(y_test, predictions)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, predictions))
```

```python
model_results("EasyEnsembleClassifier",y_test,predictions)
results_dict["EasyEnsembleClassifier"]['classification_report_imbalanced']
```
